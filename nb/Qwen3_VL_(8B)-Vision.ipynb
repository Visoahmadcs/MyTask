```python
import pandas as pd
from tqdm import tqdm

def run_inference_and_export(model, tokenizer, dataset, output_path, start_idx=0, end_idx=None):
    """
    Run VQA inference on dataset and export results to Excel format compatible with evaluation script.

    Args:
        model: Your VQA model
        tokenizer: Model tokenizer
        dataset: Dataset (brain_ds in your case)
        output_path: Path to save Excel file
        start_idx: Starting index for dataset slice
        end_idx: Ending index for dataset slice (None for full dataset)
    """

    if end_idx is None:
        end_idx = len(dataset)

    results = []

    for i in tqdm(range(start_idx, end_idx), desc="Running VQA inference"):
        try:
            # Get data
            image = dataset[i]["image"]
            #questions =   dataset[i]["Question"]
            #instruction = "Answer in this EXACT format: [Plane: sagittal/axial/coronal], [Tumor: Glioma/Meningioma/Pituitary/No]"
            #instruction = questions + "? Analyze in detail."
            #instruction = random.choice(QUESTION_TEMPLATES)
            #instruction = "Provide only the bounding box coordinates for the detected abnormality in this brain MRI scan. Format the output as follows: bbox: []"
            # instruction = """You are a medical imaging AI. Analyze the brain MRI and detect the tumor.

            #             **OUTPUT REQUIREMENTS (FOLLOW STRICTLY):**
            #             1. **Coordinate System**: Return normalized floats in range [0.0, 1.0]
            #             - Divide pixel coordinates by image width and height
            #             2. **Format**: xyxy bounding box: [x_min, y_min, x_max, y_max]
            #             3. **Precision**: Exactly 6 decimal places
            #             4. **Output**: Single-line, no extra text: bbox: [0.453125, 0.375000, 0.578125, 0.484375]

            #             **Tumor Identification:**
            #             - Glioma appears hyperintense (bright) on T1-weighted MRI
            #             - Focus on largest abnormal mass
            #             - Avoid ventricles and normal tissue

            #             Now analyze and output bbox:"""

            instruction =  "Describe this Brain MRI image. Include tumor type, location, and MRI plane."

            gt_answer = dataset[i]["caption"]
            #bbox = dataset[i]["bbox"]

            # Prepare messages
            messages = [
                {"role": "user", "content": [
                    {"type": "image"},
                    {"type": "text", "text": instruction}
                ]}
            ]

            # Tokenize
            input_text = tokenizer.apply_chat_template(messages, add_generation_prompt=True)
            inputs = tokenizer(image, input_text, add_special_tokens=False, return_tensors="pt").to("cuda")

            # Generate
            with torch.no_grad():
                outputs = model.generate(
                    **inputs,
                    max_new_tokens=50,
                    use_cache=True,
                    temperature=0.7,
                    top_p=0.9,
                    do_sample=False,
                    pad_token_id=tokenizer.eos_token_id
                )

            # Decode result
            generated_answer = tokenizer.decode(
                outputs[0][inputs['input_ids'].shape[1]:],
                skip_special_tokens=True
            ).strip()

            # Store results
            results.append({
                'id': i,  # This will be used as image_id in evaluation
                'gt': gt_answer,  # Ground truth answer
                'parsed_output': generated_answer,  # Model prediction
                #'bbox':bbox,
                #'question': instruction  # Optional: keep question for reference
            })

        except Exception as e:
            print(f"Error processing sample {i}: {e}")
            # Add placeholder for failed samples
            results.append({
                'id': i,
                'gt': dataset[i]["caption"] if i < len(dataset) else "ERROR",
                'parsed_output': "ERROR",
                #'question': dataset[i]["Question"] if i < len(dataset) else "ERROR"
            })

    # Create DataFrame and save
    df = pd.DataFrame(results)
    df.to_excel(output_path, index=False)
    print(f"Saved {len(results)} results to {output_path}")

    return df
```

```python
df = run_inference_and_export(
    model=model,
    tokenizer=tokenizer,
    dataset=dataset,
    output_path='./joint_model_ic_results.xlsx',
    start_idx=0,
    end_idx=100  # Process first 100 samples
)
```

```python
# ============= MINIMAL BATCHED VQA =============
import torch
from tqdm import tqdm
import pandas as pd

# FastVisionModel.for_inference(model)
# model.eval()

BATCH_SIZE = 24  # Adjust based on GPU memory
results = []

# Pre-extract all questions for batching
all_questions = []
all_images = []
all_gts = []
all_ids = []

for i, sample in enumerate(dataset):
    for qa in sample.get("vqa", []):
        all_images.append(sample["image"])
        all_questions.append(qa["question"])
        all_gts.append(qa["answer"])
        all_ids.append(i)

print(f"Processing {len(all_questions)} questions...")

# Batch inference
for batch_start in tqdm(range(0, len(all_questions), BATCH_SIZE), desc="VQA"):
    batch_end = min(batch_start + BATCH_SIZE, len(all_questions))

    # Build batch
    batch_images = all_images[batch_start:batch_end]
    batch_questions = all_questions[batch_start:batch_end]

    # Tokenize batch
    inputs = tokenizer(
        images=batch_images,
        text=[tokenizer.apply_chat_template(
            [{"role": "user", "content": [{"type": "image"}, {"type": "text", "text": q}]}],
            add_generation_prompt=True
        ) for q in batch_questions],
        return_tensors="pt",
        padding=True,
        truncation=True,
        add_special_tokens=False
    ).to("cuda")

    # Generate (greedy for speed)
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=32,  # Answers are short
            do_sample=False,     # Greedy decoding is faster & deterministic
            use_cache=True,
            pad_token_id=tokenizer.eos_token_id
        )

    # Decode
    input_lengths = inputs['input_ids'].shape[1]
    predictions = tokenizer.batch_decode(
        outputs[:, input_lengths:],
        skip_special_tokens=True
    )

    # Store results
    for j, pred in enumerate(predictions):
        idx = batch_start + j
        results.append({
            'id': all_ids[idx],
            'gt': all_gts[idx],
            'parsed_output': pred.strip(),
            'question': all_questions[idx]
        })

# Save
df = pd.DataFrame(results)
df.to_excel("vqa_baseline_results.xlsx", index=False)
print(f"‚úÖ Saved {len(results)} results")
```

```python
instruction = "Write the LaTeX representation for this image."

def convert_to_conversation(sample):
    conversation = [
        { "role": "user",
          "content" : [
            {"type" : "text",  "text"  : instruction},
            {"type" : "image", "image" : sample["image"]} ]
        },
        { "role" : "assistant",
          "content" : [
            {"type" : "text",  "text"  : sample["text"]} ]
        },
    ]
    return { "messages" : conversation }
pass
```

### Train the model
Now let's train our model. We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`. We also support TRL's `DPOTrainer`!

We use our new `UnslothVisionDataCollator` which will help in our vision finetuning setup.

```python
from unsloth.trainer import UnslothVisionDataCollator
from trl import SFTTrainer, SFTConfig

FastVisionModel.for_training(model) # Enable for training!

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    data_collator = UnslothVisionDataCollator(model, tokenizer), # Must use!
    train_dataset = converted_dataset,
    args = SFTConfig(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 30,
        # num_train_epochs = 1, # Set this instead of max_steps for full training runs
        learning_rate = 2e-4,
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.001,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none",     # For Weights and Biases

        # You MUST put the below items for vision finetuning:
        remove_unused_columns = False,
        dataset_text_field = "",
        dataset_kwargs = {"skip_prepare_dataset": True},
        max_length = 2048,
    ),
)
```

```python
# @title Show current memory stats
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
```

```python
trainer_stats = trainer.train()
```

```python
# @title Show final memory and time stats
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory / max_memory * 100, 3)
lora_percentage = round(used_memory_for_lora / max_memory * 100, 3)
print(f"{trainer_stats.metrics['train_runtime']} seconds used for training.")
print(
    f"{round(trainer_stats.metrics['train_runtime']/60, 2)} minutes used for training."
)
print(f"Peak reserved memory = {used_memory} GB.")
print(f"Peak reserved memory for training = {used_memory_for_lora} GB.")
print(f"Peak reserved memory % of max memory = {used_percentage} %.")
print(f"Peak reserved memory for training % of max memory = {lora_percentage} %.")
```

### GGUF / llama.cpp Conversion
To save to `GGUF` / `llama.cpp`, we support it natively now! We clone `llama.cpp` and we default save it to `q8_0`. We allow all methods like `q4_k_m`. Use `save_pretrained_gguf` for local saving and `push_to_hub_gguf` for uploading to HF.

Some supported quant methods (full list on our [Wiki page](https://github.com/unslothai/unsloth/wiki#gguf-quantization-options)):
* `q8_0` - Fast conversion. High resource use, but generally acceptable.
* `q4_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q4_K.
* `q5_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q5_K.

[**NEW**] To finetune and auto export to Ollama, try our [Ollama notebook](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Llama3_(8B)-Ollama.ipynb)
